#  Algorithm 1 (CountVectorizer & Cosine Similarity)

## CountVectorizer 

### Create a CountVectorizer

```python
#Create vectorizer
cv = CountVectorizer(analyzer='word', lowercase=True, stop_words = 'english',ngram_range = (1,2),min_df = 0.002)
#ngram_range = (1,2) - if used more than  1(value), lots of features or noise. It wil be ignore when call analyzer
#min_df = rare words, max_df = most used words
```

### Fit the keyword column with cv

```python
#Fit data into vectorizer
cv.fit(df_book['keywords'])
```

### Map the vocabulary with index in the cv

Note: Machine cannot understand the vocabulary, but can understand each index is diff meaning of the word

```python
#The sorting happens based on ASCII sort order.
display("Total number of vocabulary = " + str(len(cv.vocabulary_)))
display("Data type = " + str(type(cv.vocabulary_)))
display("Display All of the vocabulary and index")
display(cv.vocabulary_)
```

### Display all of the vocabulary in the cv

```python
display("The vocabulary that in the tfid")
display(cv.get_feature_names())
```

### Create a matrix based on the column

```python
matrix_cv = cv.fit_transform(df_book['keywords'])
```

### Display matrix

```python
print("Row = Keyword or record, Column = importance & frequency of the vocabulary")
print(f"(number of rows, number of Columns) = {matrix_cv.shape}")
display("Dimension of matrix = " + str(matrix_cv.ndim))
display("Data type = " + str(type(matrix_cv)))
print(matrix_cv)
```

## Cosine Similarity

### Calculate the cosine similarity between two matrix

```python
cosine_sim_cv = cosine_similarity(matrix_cv,matrix_cv)
```

### Display the matrix

```python
print(f"(number of rows, number of Columns) = {cosine_sim_cv.shape}")
display("Data type = " + str(type(cosine_sim_cv)))
display("Dimension of matrix = " + str(cosine_sim_cv.ndim))
display(cosine_sim_cv)
```

## Find the user input

### Create a series

```python
series_cv = pd.Series(df_book.index, index = df_book['title'])
display(series_cv)
```

### FInd the index based on the input by user and search from series

```python
user_book_id_cv = series_cv[input_book_name]
user_book_id_cv
```

### Check the result

```python
#Display the title only
display("Title: " + df_book['title'].iloc[user_book_id_cv])
display("Relevant details of " + df_book['title'].iloc[user_book_id_cv])
print("--------------------------")
display(df_book.iloc[user_book_id_cv])
```

###  Find out what features have been considered  by the model for a given keyword

For the title "Harry Potter and the Philosopher's Stone", the feature array might include words like "harry", "potter", "philosopher", "stone", "hogwarts", "j.k.", "rowling", "magic", etc.
<br>This feature array we are using keyword not title, title just an example.
<br>Keyword =title + tag

```python
#Convert matrix to 1D which refer to array
feature_array_cv = np.squeeze(matrix_cv[user_book_id_cv].toarray()) 
display(feature_array_cv.shape)
display("Data type = " + str(type(feature_array_cv)))
display("Dimension of matrix = " + str(feature_array_cv.ndim))
display(  + feature_array_cv.size) 
print(feature_array_cv)
```

## Find the value of feature array more than 0

```python
#Get the vocabulary from count vectorizer
vocabulary_cv = cv.get_feature_names()
#Array value = frequency of word
#Get the index for the value more than 0
idx_cv = np.where(feature_array_cv > 0)

display(idx_cv[0])
```

### Print the related vocabulary for the value more than 0

```python
print([vocabulary_cv[x] for x in idx_cv[0]])
```

### Sort the value and get the index

```python
n = input_book_recommend_count #how many books to be recommended

#np.argsort = ascending order
#flip = inverse the order so can get the descending order
top_idx_cv = np.flip(np.argsort(cosine_sim_cv[user_book_id_cv,]), axis = 0)[0:n]
display(top_idx_cv)
display("Display the relevant book")
display(df_book['title'].iloc[top_idx_cv])
```

##  Calculate the cosine similarity scores of similar books

```python
top_sim_values_cv = cosine_sim_cv[user_book_id_cv, top_idx_cv]
top_sim_values_cv
```

## Find the index and score of the top similarity

```python
#find top n with values > 0
top_idx_cv = top_idx_cv[top_sim_values_cv  > 0] ## Index
scores_cv = top_sim_values_cv[top_sim_values_cv  > 0] ## Score
display("Index of similar book") 
display(top_idx_cv)
display("Score of similar book")
display(scores_cv)
```

## Result for this model


```python
#set column width to display the complete title
pd.options.display.max_colwidth = 150
#display the book title and its similarity value in dataframe
pd.DataFrame({"Recommended Books": df_book['title'].iloc[top_idx_cv].values,
              "Scores":scores_cv}, columns = ["Recommended Books","Scores"])
```
